````python
omega_21 = 5.9e-6 * eV
eps_baseline = 1e-6
omega_baseline = omega_21
tau_u = 13.8e9 * Year
````

````python
def plot_m_A(m_A, z_ary, plot_m_A=True):
    z_cross, P_mult = spec.P_tot(omega_0=omega_baseline, m_A=m_A, eps=eps_baseline, evolve_z=False)
    
    if plot_m_A:
        plt.plot(z_ary, [np.sqrt(spec.m_A_sq(z=z, omega=omega_21)) / eV for z in z_ary])
        plt.axhline(m_A / eV)
        for z in z_cross:
            plt.axvline(z)
        plt.xscale("log")
        plt.yscale("log")
        
````

````python
def get_max_CMB_photon_ratio_z_17(f_DM, z_res, m_a, tau_a, T_d_recomb=None):
    tau_a_div_tau_u = tau_a / tau_u
    m_A = np.sqrt(spec.m_A_sq(z=z_res, omega=omega_21))
    
    z_cross, P_mult = spec.P_tot(omega_baseline, eps_baseline, m_A, evolve_z=False)
        
    z_P_bin_ary = np.logspace(-1,3.2,1000)
    z_P_bin_centers_ary = 10 ** ((np.log10(z_P_bin_ary[1:]) + np.log10(z_P_bin_ary[:-1])) / 2.)
    P_ary = np.zeros(len(z_P_bin_ary) - 1)
        
    for i_z, z in enumerate(z_cross):
        P_ary[np.where(z_P_bin_ary < z)[0][-1]] = P_mult[i_z]

    # Checks against existing constraints

    if m_a / 2 / (1 + z_cross[0]) < omega_radio_ary[0]:
        eps_ARCADE2 = 99.
    else:
        eps_ARCADE2 = eps_baseline * np.sqrt(spec.ratio_ARCADE2(m_a, z_P_bin_centers_ary, P_ary, tau_a, f_DM=f_DM, T_d_recomb=T_d_recomb))

    f_a = spec.f_a(tau_a, m_a)

    eps_stellar = spec.eps_stellar_cooling(f_a)
    eps_max_FIRAS = 10 ** np.interp(np.log10(m_A), np.log10(m_Ap_FIRAS), np.log10(eps_bound_FIRAS))
    
    eps_max = np.min([eps_stellar, eps_ARCADE2, eps_max_FIRAS])
        
    # Get ratio of 21-cm photon number densities at z=17
    dn_CMB_domega = spec.dn_CMB_domega(omega=omega_21, z=17)
    dn_A_domega = spec.dn_A_domega(m_a=m_a, z=17, z_res_ary=z_P_bin_centers_ary, P_ary=P_ary * (eps_max / eps_baseline) ** 2,
                                                     omega=omega_21, tau_a=tau_a, f_DM=f_DM, T_d_recomb=T_d_recomb)
    return m_A, eps_max, z_P_bin_centers_ary, z_P_bin_ary, P_ary, ((dn_A_domega + dn_CMB_domega) / dn_CMB_domega)
````

````python
def as_si(x, ndp):
    s = '{x:0.{ndp:d}e}'.format(x=x, ndp=ndp)
    m, e = s.split('e')
    if float(m) == 1.:
        return r'10^{{{e:d}}}'.format(m=m, e=int(e))
    else:
        return r'{m:s}\times 10^{{{e:d}}}'.format(m=m, e=int(e))
````

# II. Stimulated emission

$$
\Delta N_{\mathrm{eff}}=\frac{8}{7}\left(\frac{11}{4}\right)^{4 / 3}\left.\frac{\rho_{d}}{\rho_{\gamma}}\right|_\mathrm{recomb.}
$$
$\rho_{\gamma}$ is the CMB energy density at recombination, $\rho_{d}$ is energy density of dark photons (characterized by a temperature $T_d$) at recombination time.

````python
z_recomb = 1089.80 # Recombination temperature, Planck 2018
T_CMB_recomb = spec.T_CMB(z=z_recomb)
Delta_Neff = 0.3 # Limit on Delta N_eff, ~dummy for now
T_d_recomb = (Delta_Neff / ((8 / 7.) * (11 / 4.) ** (4/3.)) * T_CMB_recomb ** 4) ** (1 / 4.)

f_DM = 1.
z_res = 130.
m_a = 10 ** -4. * eV

tau_a = spec.tau_a_LL() 
tau_a_div_tau_u = tau_a / tau_u

m_A, eps_max, z_P_bin_centers_ary,z_P_bin_ary, P_ary, ratio_z_17 = get_max_CMB_photon_ratio_z_17(f_DM, z_res, m_a, tau_a, T_d_recomb=T_d_recomb)
print(ratio_z_17)
````
