```python
df_all = pd.read_csv('../data/apjlaabf86t2_ascii.txt', delimiter='\t', skiprows=list(range(6)) + list(range(21,24)), names=['freq', 'survey', 'fid', 'unc', 'n/a'])
df_all.head()

df_arcade2 = df_all[df_all['survey'].str.contains("ARCADE")]
df_llfss = df_all[df_all['survey'].str.contains("LLFSS")]
df_radio = df_all[~df_all['survey'].str.contains("LLFSS|ARCADE")]

omega_ARCADE2_ary = 2 * np.pi * df_arcade2['freq'].values * 1e9 / Sec
T_ARCADE2_fid = df_arcade2['fid'].values
T_ARCADE2_unc = df_arcade2['unc'].values

omega_radio_ary = 2 * np.pi * df_radio['freq'].values * 1e9 / Sec
T_radio_fid = df_radio['fid'].values
T_radio_unc = df_radio['unc'].values

omega_llfss_ary = 2 * np.pi * df_llfss['freq'].values * 1e9 / Sec
T_llfss_fid = df_llfss['fid'].values
T_llfss_unc = df_llfss['unc'].values
```

Eq. (7) of 1902.02438:
$$
T_{\mathrm{rad}}=T_{\mathrm{CMB}}(1+z)\left[1+A_{\mathrm{r}}\left(\frac{\nu_{\mathrm{obs}}}{78 \mathrm{MHz}}\right)^{\beta}\right]
$$

```python
def T_CMB_mod(A_r, beta, T_CMB_0, omega_base, omega_obs):
    return T_CMB_0 * (1 + A_r * (omega_obs / omega_base) ** beta)
```

```python
omega_ary = np.concatenate((omega_ARCADE2_ary / eV, omega_radio_ary / eV, omega_llfss_ary / eV))
T_ary = np.concatenate((T_ARCADE2_fid, T_radio_fid, T_llfss_fid))
T_unc_ary = np.concatenate((T_ARCADE2_unc, T_radio_unc, T_llfss_unc))

def chi2(x):
    A_r, beta, T_CMB_0 = x
    return np.sum(((T_CMB_mod(A_r, beta, T_CMB_0, omega_base_0, omega_ary) - T_ary) / T_unc_ary) ** 2)

from scipy.optimize import minimize

omega_base_0 = 2 * np.pi * 78 * 1e6 / Sec / eV  # Observed frequency in eV

opt = minimize(chi2, x0=[100, -2.6, 2.725])
opt
```

```python
omega_plot_ary = np.logspace(np.log10(np.min(omega_ary)), np.log10(np.max(omega_ary)))
T_CMB_mod_plot_ary = T_CMB_mod(*opt.x, omega_base_0, omega_plot_ary)
```

```python
plt.plot(omega_plot_ary, T_CMB_mod_plot_ary, label="Fit")

plt.errorbar(omega_ARCADE2_ary / eV, T_ARCADE2_fid, yerr=T_ARCADE2_unc,
            fmt='o', capthick=1, capsize=1, markersize=0, color='magenta',
            label="ARCADE")

plt.errorbar(omega_radio_ary / eV, T_radio_fid, yerr=T_radio_unc,
            fmt='o', capthick=1, capsize=1, markersize=0, color='blue',
            label="Other radio")

plt.errorbar(omega_llfss_ary / eV, T_llfss_fid, yerr=T_llfss_unc,
            fmt='o', capthick=1, capsize=1, markersize=0, color='green',
            label="LWA1")

plt.xscale("log")
plt.yscale("log")

plt.ylabel("$T_\mathrm{CMB}\,[\mathrm K]$")
plt.xlabel(r"$\omega$\,[eV]")
plt.xscale("log")
plt.yscale("log")

plt.ylabel("$T_\mathrm{CMB}\,[\mathrm K]$")
plt.xlabel(r"$\omega$\,[eV]")

plt.legend()
```

```python
A_r_fit, beta_fit, T_CMB_0_fit = opt.x
```

```python
fraction_excess = 1e-2 / 2  # What fraction of excess radiation contributes?
omega_21 = 5.9e-6 * eV

z_ary = np.linspace(5,50)

z_bench = 17

print((T_CMB_mod(fraction_excess * A_r_fit, beta_fit, T_CMB_0_fit, omega_base_0, omega_21 / eV / (1 + z_bench)) * (1 + z_bench)) / (T_CMB_0_fit * (1 + z_bench)))

T_CMB_mod_ary = T_CMB_mod(fraction_excess * A_r_fit, beta_fit, T_CMB_0_fit, omega_base_0, omega_21 / eV / (1 + z_ary)) * (1 + z_ary)

plt.plot(z_ary, T_CMB_mod_ary)
plt.plot(z_ary, T_CMB_0_fit * (1 + z_ary))

plt.yscale("log")

plt.xlabel("$z$")
plt.ylabel("$T_\mathrm{CMB}\,[\mathrm K]$")
```
