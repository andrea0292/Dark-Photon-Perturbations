## 1. Power spectrum from `CLASS`

```python
# Cosmology from Hongwan

z_compute = 20.

common_settings = {'output':'mTk,mPk,vTk',
                   'omega_b':0.022032,'omega_cdm':0.12038,'h':0.67556,'A_s':2.215e-9,'n_s':0.9619,'tau_reio':0.0925,
                   'P_k_max_1/Mpc':100.0,
                   'non linear': 'halofit'}

M = Class()
M.set(common_settings)
M.set({'z_pk': 1.088e+03})
M.compute()
one_time = M.get_transfer(z_compute, 'camb')

kk = np.logspace(-4,np.log10(3),1000) # k in h/Mpc
Pk = [] # P(k) in (Mpc/h)**3
h = M.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk:
    Pk.append(M.pk(k*h,z_compute)) # function .pk(k,z)
    
# Transfer functions
k_ary = one_time['k (h/Mpc)'] * h
T_ary = one_time['-T_b/k2']
T_m_ary = one_time['-T_tot/k2']

n_s = M.n_s()
P_s = 2.215e-9 * k_ary ** (n_s - 1)  # Primordial PS 

# Power spectra from transfer function
Pk_fromt = P_s * T_ary ** 2 * k_ary ** 4 / (k_ary ** 3 / (2 * np.pi ** 2)) * h **3
Pkm_fromt = P_s * T_m_ary ** 2 * k_ary ** 4 / (k_ary ** 3 / (2 * np.pi ** 2)) * h **3
```

```python
plt.plot(k_ary / h, Pk_fromt, label='Baryon PS')
plt.plot(k_ary / h, Pkm_fromt, label='Matter PS')

# plt.plot(kk * h, np.array(Pk))

plt.xscale("log")
plt.yscale("log")

# plt.ylim(1e-15,1e1)
plt.xlim(1e-6,3e3)

plt.ylabel("$P_\mathrm{b}(k)$\,[Mpc$^3$/$h^3$]")
plt.xlabel("$k$\,[$h$/Mpc]")

plt.title("$z=" + str(z_compute) + "$")

plt.legend()
```

```python
# -*- coding: utf-8 -*-

def generate_field(statistic, power_spectrum, shape, unit_length=1, filt_size=1):
    """
    Generates a field given a stastitic and a power_spectrum.
    Parameters
    ----------
    statistic: callable
        A function that takes returns a random array of a given signature,
        with signature (s) -> (B) with B.shape == s. Please note that the
        distribution is in *Fourier space* not in real space, unless you set
        stat_real=True. See the note below for more details.
    power_spectrum: callable
        A function that returns the power contained in a given mode,
        with signature (k) -> P(k) with k.shape == (ndim, n)
    shape: tuple
        The shape of the output field
    unit_length: float, optional
        How much physical length represent 1pixel. For example a value of 10
        mean that each pixel stands for 10 physical units. It has the
        dimension of a physical_unit/pixel.
    Returns
    -------
    field: a real array of shape `shape` following the statistic
        with the given power_spectrum
    Note
    ----
    When generation the distribution in Fourier mode, the result
    should be complex and unitary. Only the phase is random.
    """
    
    # Compute the k grid
    all_k = [np.fft.fftfreq(s, d=unit_length) for s in shape[:-1]] + \
            [np.fft.rfftfreq(shape[-1], d=unit_length)]

    kgrid = np.meshgrid(*all_k, indexing='ij')
    knorm = 2*np.pi*np.sqrt(np.sum(np.power(kgrid, 2), axis=0))
    
    fourier_shape = knorm.shape
    
    vol = np.prod(np.array(shape) * unit_length)
    scale_factor = vol

    # Draw a random sample in Fourier space
#     pm = ParticleMesh(Nmesh=shape, BoxSize=shape * unit_length) 
#     fftfield = np.array(pm.generate_whitenoise(seed=55,unitary=1)) #
    fftfield = statistic(fourier_shape)

    power_k = np.where(knorm == 0, 0, np.sqrt(power_spectrum(knorm)/scale_factor))
    tophat_k = np.where(knorm == 0, 0, tophat(filt_size, knorm))
    fftfield *= tophat_k * power_k
    return np.prod(shape)*np.fft.irfftn(fftfield)

def distrib(shape):
    # Build a unit-distribution of complex numbers with random phase
    a = np.random.normal(loc=0, scale=1, size=shape)
    b = np.random.normal(loc=0, scale=1, size=shape)
    return (a + 1j * b)/np.sqrt(2)

# def distrib(shape):
#     # Build a unit-distribution of complex numbers with random phase
#     phi = np.random.uniform(0, 2 * np.pi, size=shape)
#     return (np.cos(phi) + 1j * np.sin(phi))


def distrib_real(shape):
    # Build a unit-distribution of complex numbers with random phase
    a = np.random.normal(loc=0, scale=1, size=shape)
    return a
```

```python
def tophat(r, knorm):
    knorm = np.array(knorm)
#     r = 1.
#     k = np.sum(ki ** 2 for ki in k) ** 0.5
    kr = knorm * r
#     print(kr)
    w = 3 * (np.sin(kr) / kr **3 - np.cos(kr) / kr ** 2)
#     print(w)
    w[k == 0] = 1.0
    return w
```
