```python
h1_picke = np.load("../data/redshift_ion_heat_eV_rate_dz_decaying_mDM_1e7_tau_1e25.p", allow_pickle=True)

one_plus_z = h1_picke[0]
dx_e_dz = h1_picke[1]
dT_dz = h1_picke[2] * eV / Kelv

z_ary = np.logspace(np.log10(5), np.log10(2900), 100)
z_max = 2900

dx_e_dz_interp = interp1d(one_plus_z - 1, dx_e_dz, bounds_error=False, fill_value="extrapolate")
dx_e = np.array([quad(lambda z: dx_e_dz_interp(z), z_max, z) for z in z_ary])[:, 0]

dT_b_dz_interp = interp1d(one_plus_z - 1, dT_dz, bounds_error=False, fill_value="extrapolate")
dT_b = np.array([quad(lambda z: dT_b_dz_interp(z), z_max, z) for z in z_ary])[:, 0]

dx_e_interp = interp1d(z_ary, dx_e, fill_value="extrapolate")
dT_b_interp = interp1d(z_ary, dT_b, fill_value="extrapolate")
```

```python
h1_picke_alt = np.load("../data/redshift_ion_heat_eV_rate_dz_decaying_mDM_1e7_tau_3e25.p", allow_pickle=True)

one_plus_z_alt = h1_picke_alt[0]
dx_e_dz_alt = h1_picke_alt[1]
dT_dz_alt = h1_picke_alt[2] * eV / Kelv

# z_ary = np.logspace(np.log10(5), np.log10(2900), 100)
z_ary = np.linspace(5, 2900, 100)

z_max = 2900

dx_e_dz_interp_alt = interp1d(one_plus_z - 1, dx_e_dz_alt, bounds_error=False, fill_value="extrapolate")
dx_e_alt = np.array([quad(lambda z: dx_e_dz_interp_alt(z), z_max, z) for z in z_ary])[:, 0]

dT_b_dz_interp_alt = interp1d(one_plus_z - 1, dT_dz_alt, bounds_error=False, fill_value="extrapolate")
dT_b_alt = np.array([quad(lambda z: dT_b_dz_interp_alt(z), z_max, z) for z in z_ary])[:, 0]

dx_e_interp_alt = interp1d(z_ary, dx_e_alt, fill_value="extrapolate")
dT_b_interp_alt = interp1d(z_ary, dT_b_alt, fill_value="extrapolate")
```
