# I. Parameter scans (fiducial case)

```python
eps_baseline = 1e-6
omega_21 = 5.9e-6 * eV
tau_u = 13.8e9 * Year
```

```python
m_Ap_FIRAS = np.load('../data/FIRAS_bounds.npz')['m_Ap'] * eV
eps_bound_FIRAS = np.load('../data/FIRAS_bounds.npz')['eps_max']
```

```python
df_all = pd.read_csv('../data/apjlaabf86t2_ascii.txt', delimiter='\t', skiprows=list(range(6)) + list(range(21,24)), names=['freq', 'survey', 'fid', 'unc', 'n/a'])
df_all.head()

df_arcade2 = df_all[df_all['survey'].str.contains("ARCADE")]
df_llfss = df_all[df_all['survey'].str.contains("LLFSS")]
df_radio = df_all[~df_all['survey'].str.contains("LLFSS|ARCADE")]

omega_ARCADE2_ary = 2 * np.pi * df_arcade2['freq'].values * 1e9 / Sec
T_ARCADE2_fid = df_arcade2['fid'].values
T_ARCADE2_unc = df_arcade2['unc'].values

omega_radio_ary = 2 * np.pi * df_radio['freq'].values * 1e9 / Sec
T_radio_fid = df_radio['fid'].values
T_radio_unc = df_radio['unc'].values

omega_llfss_ary = 2 * np.pi * df_llfss['freq'].values * 1e9 / Sec
T_llfss_fid = df_llfss['fid'].values
T_llfss_unc = df_llfss['unc'].values
```

```python
m_A_ary = np.logspace(-14, -9, 100) * eV
m_a_ary = np.logspace(-5, -1, 50) * eV

m_a_EDGES_ary = np.zeros(len(m_A_ary))

ratio_ary = np.zeros((len(m_A_ary), len(m_a_ary)))

spec = Spectra()
tau_a = spec.tau_a_LL() 

omega_baseline = 5.9e-6 * eV
eps_baseline = 1e-6

for i_m_A, m_A in enumerate(tqdm(m_A_ary)):
    
    z_cross, P_mult = spec.P_tot(omega_baseline, eps_baseline, m_A, evolve_z=False)
        
    if len(z_cross) == 0:
        m_a_EDGES_ary[i_m_A] = 0.
        ratio_ary[i_m_A, :] = 0. 
        continue
        
    z_P_bin_ary = np.logspace(-2, 4,1000)
    P_ary = np.zeros(len(z_P_bin_ary) - 1)

    for i_z, z in enumerate(z_cross):
        P_ary[np.where(z_P_bin_ary < z)[0][-1]] = P_mult[i_z]

    z_res = z_cross[-1]

    # Redshift bins at which transition probabilities specified
    z_P_bin_centers_ary = 10 ** ((np.log10(z_P_bin_ary[1:]) + np.log10(z_P_bin_ary[:-1])) / 2.)

    m_a_EDGES  = 2 * omega_21 / ((15. + 1) / (1 + z_res))
    m_a_EDGES_ary[i_m_A] = m_a_EDGES
    
    for i_m_a, m_a in enumerate(m_a_ary):

        if m_a / 2 / (1 + z_cross[0]) < omega_radio_ary[0]:
            eps_ARCADE2 = 99.
        else:
            eps_ARCADE2 = eps_baseline * np.sqrt(spec.ratio_ARCADE2(m_a, z_P_bin_centers_ary, P_ary, tau_a))

        f_a = spec.f_a(tau_a, m_a)
        
        eps_stellar = spec.eps_stellar_cooling(f_a)

        eps_max_FIRAS = 10 ** np.interp(np.log10(m_A), np.log10(m_Ap_FIRAS), np.log10(eps_bound_FIRAS))

        eps_max = np.min([eps_stellar, eps_ARCADE2, eps_max_FIRAS])

        dn_CMB_domega = spec.dn_CMB_domega(omega=omega_21, z=17)
        dn_A_domega = spec.dn_A_domega(m_a=m_a, z=17, z_res_ary=z_P_bin_centers_ary, P_ary=P_ary *  (eps_max / eps_baseline) ** 2,
                                                         omega=omega_21, tau_a=tau_a)

        ratio_ary[i_m_A, i_m_a] = dn_A_domega / dn_CMB_domega + 1
```

```python
z_res_plot_ary = [20, 40, 50, 80, 100, 130]
m_A_plot_ary = []
for z_res in z_res_plot_ary:
    m_A = np.sqrt(spec.m_A_sq(z=z_res, omega=omega_21)) 
    m_A_plot_ary.append(m_A / eV)
```

```python
mpl.rcParams['figure.figsize'] = 7.5, 5.5

ratio_ary_trans = np.transpose(ratio_ary)

levels = [3., 5.]
X, Y = np.meshgrid(np.log10(m_A_ary / eV), np.log10(m_a_ary / eV))

fig, ax = plt.subplots()

CS = ax.contour(X, Y, ratio_ary_trans, levels = levels, linestyles=['-','--'], colors='k', linewidths=2, alpha=0.7)

fmt = {}
strs = ['3', '5']
for l, s in zip(CS.levels, strs):
    fmt[l] = s
    
im = ax.imshow(np.log10(ratio_ary_trans), cmap = 'YlGn', aspect='auto', origin='lower', extent=[-14, -9, -5, -1])

ax.clabel(CS, fmt=fmt,inline=1, fontsize=14)

ax.plot(np.log10(m_A_ary / eV), np.log10(m_a_EDGES_ary / eV), color='blue', label='$z_\mathrm{end} = 15$')

cbar = fig.colorbar(im)
cbar.set_label(r'$\log_{10}(T_A^{\mathrm{max}} / T_{\mathrm{CMB}})_{z=17}$')

ax.set_xlabel('$\log_{10}\,(m_A^\prime\,[\mathrm{eV}])$')
ax.set_ylabel('$\log_{10}\,(m_a\,[\mathrm{eV}])$')

ax.set_xlim(-14, -9)
ax.set_ylim(-6, -1)

# ax_top = ax.twiny()
# ax_top.set_xlim(1,2)

colors = ['red', 'green', 'orange', 'magenta', 'grey', 'pink']
for i_z_res, z_res in enumerate(z_res_plot_ary):
    m_A = np.sqrt(spec.m_A_sq(z=z_res, omega=omega_21)) 
    plt.axvline(np.log10(m_A / eV), label='$z_\mathrm{res} = ' + str(z_res_plot_ary[i_z_res]) + "$", color = colors[i_z_res])

plt.legend(loc='lower right', ncol=2)

plt.title(r"\bf{Maximum allowed injection at $z=17$}")

plt.tight_layout()

# plt.savefig("../draft/plots/parameter_space.pdf")
```

# II. Parameter scans (stimulated emission saturating $N_\mathrm{eff}$ bounds)

```python
m_A_ary = np.logspace(-14, -9, 100) * eV
m_a_ary = np.logspace(-5, -1, 50) * eV

m_a_EDGES_ary = np.zeros(len(m_A_ary))

ratio_ary = np.zeros((len(m_A_ary), len(m_a_ary)))

spec = Spectra()
tau_a = spec.tau_a_LL() 

omega_baseline = 5.9e-6 * eV
eps_baseline = 1e-6

for i_m_A, m_A in enumerate(tqdm(m_A_ary)):
    
    z_cross, P_mult = spec.P_tot(omega_baseline, eps_baseline, m_A, evolve_z=False)
        
    if len(z_cross) == 0:
        m_a_EDGES_ary[i_m_A] = 0.
        ratio_ary[i_m_A, :] = 0. 
        continue
        
    z_P_bin_ary = np.logspace(-1,4,1000)
    P_ary = np.zeros(len(z_P_bin_ary) - 1)

    for i_z, z in enumerate(z_cross):
        P_ary[np.where(z_P_bin_ary < z)[0][-1]] = P_mult[i_z]

    z_res = z_cross[-1]

    # Redshift bins at which transition probabilities specified
    z_P_bin_centers_ary = 10 ** ((np.log10(z_P_bin_ary[1:]) + np.log10(z_P_bin_ary[:-1])) / 2.)

    m_a_EDGES  = 2 * omega_21 / ((15. + 1) / (1 + z_res))
    m_a_EDGES_ary[i_m_A] = m_a_EDGES
    
    for i_m_a, m_a in enumerate(m_a_ary):
        
        z_recomb = 1089.80 # Recombination temperature, Planck 2018
        T_CMB_recomb = spec.T_CMB(z=z_recomb)
        Delta_Neff = 0.3 # Limit on Delta N_eff, ~dummy for now
        T_d_recomb = (Delta_Neff / ((8 / 7.) * (11 / 4.) ** (4/3.)) * T_CMB_recomb ** 4) ** (1 / 4.)

        if m_a / 2 / (1 + z_cross[0]) < omega_radio_ary[0]:
            eps_ARCADE2 = 99.
        else:
            eps_ARCADE2 = eps_baseline * np.sqrt(spec.ratio_ARCADE2(m_a, z_P_bin_centers_ary, P_ary, tau_a, T_d_recomb=T_d_recomb))

        f_a = spec.f_a(tau_a, m_a)
        
        eps_stellar = spec.eps_stellar_cooling(f_a)

        eps_max_FIRAS = 10 ** np.interp(np.log10(m_A), np.log10(m_Ap_FIRAS), np.log10(eps_bound_FIRAS))

        eps_max = np.min([eps_stellar, eps_ARCADE2, eps_max_FIRAS])

        dn_CMB_domega = spec.dn_CMB_domega(omega=omega_21, z=17)
        dn_A_domega = spec.dn_A_domega(m_a=m_a, z=17, z_res_ary=z_P_bin_centers_ary, P_ary=P_ary *  (eps_max / eps_baseline) ** 2,
                                                         omega=omega_21, tau_a=tau_a, T_d_recomb=T_d_recomb)

        ratio_ary[i_m_A, i_m_a] = dn_A_domega / dn_CMB_domega + 1
```

```python
mpl.rcParams['figure.figsize'] = 7, 5.5

ratio_ary_trans = np.transpose(ratio_ary)

levels = [3., 5.]
X, Y = np.meshgrid(np.log10(m_A_ary / eV), np.log10(m_a_ary / eV))

fig, ax = plt.subplots()

CS = ax.contour(X, Y, ratio_ary_trans, levels = levels, linestyles=['-','--'], colors='k', linewidths=2, alpha=0.7)

fmt = {}
strs = ['3', '5']
for l, s in zip(CS.levels, strs):
    fmt[l] = s
    
im = ax.imshow(np.log10(ratio_ary_trans), cmap = 'YlGn', aspect='auto', origin='lower', extent=[-14, -9, -5, -1])

ax.clabel(CS, fmt=fmt,inline=1, fontsize=14)

ax.plot(np.log10(m_A_ary / eV), np.log10(m_a_EDGES_ary / eV), color='blue', label='$z_\mathrm{end} = 15$')

cbar = fig.colorbar(im)
cbar.set_label(r'$\log_{10}(T_A^{\mathrm{max}} / T_{\mathrm{CMB}})_{z=17}$')

ax.set_xlabel('$\log_{10}\,(m_A^\prime\,[\mathrm{eV}])$')
ax.set_ylabel('$\log_{10}\,(m_a\,[\mathrm{eV}])$')

ax.set_xlim(-14, -9)
ax.set_ylim(-6, -1)

# ax_top = ax.twiny()
# ax_top.set_xlim(1,2)

colors = ['red', 'green', 'orange', 'magenta', 'grey', 'pink']
for i_z_res, z_res in enumerate(z_res_plot_ary):
    m_A = np.sqrt(spec.m_A_sq(z=z_res, omega=omega_21)) 
    plt.axvline(np.log10(m_A / eV), label='$z_\mathrm{res} = ' + str(z_res_plot_ary[i_z_res]) + "$", color = colors[i_z_res])

plt.legend(loc='lower right', ncol=2)
plt.tight_layout()
```
